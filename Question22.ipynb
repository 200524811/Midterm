```python
class PriorityQueue:

    def __init__(self):

        self.queue = []

    def insert(self, item):

        self.queue.append(item)

    def min(self):

        return min(self.queue)

    def removeMin(self):

        self.queue.remove(self.min())

    def isEmpty(self):

        return len(self.queue) == 0

    def __len__(self):

        return len(self.queue)

    def __iter__(self):

        return self
    
    def isEmpty(self):
        return len(self.queue)==0
    
q = PriorityQueue()


q.insert(3)

q.insert(1)

q.insert(5)

q.insert(2)
print('\nThe numbers in queue are: ')

print(q.queue)


print('\nMinimum element is '),

print(q.min())

print('\nRemoving minimum element...')

q.removeMin()

print(q.queue)

print('\nMinimum element is '),

print(q.min())

print('\nLength of Queue is '),

print(q.__len__())

print('\nIs the queue empty ')
print(q.isEmpty())
```
